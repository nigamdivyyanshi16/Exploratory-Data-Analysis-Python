## Step 6: Measuring discount on a certain item

```python
data["Discount_amount"] =  data["market_price"] -data["sale_price"]
data.head()
```

```python
discount_data = data[["product","sale_price","market_price","Discount_amount"]].head(5)
```

```python
discount_data
```

```python
data["Discount_percentage"] =  (data["Discount_amount"]/data["market_price"])*100
data.head()
```

## Step 7: Find out the Missing Values from the Dataset.

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull())
```

```python
data["product"] = data["product"].fillna(data["product"].mode()[0])
```

```python
data["brand"] = data["brand"].fillna(data["brand"].mode()[0])
```

```python
data["sale_price"] = data["sale_price"].fillna(data["sale_price"].median())
```

```python
data["rating"] = data["rating"].fillna(data["rating"].median())
```

```python
data["Discount_amount"] = data["Discount_amount"].fillna(data["Discount_amount"].median())
data["Discount_percentage"] = data["Discount_percentage"].fillna(data["Discount_percentage"].median())
```

```python
sns.heatmap(data.isnull())
```

```python
data.isnull().sum()
```

## Handling Outlier

```python
data.drop("index",axis = 1, inplace = True)
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(data['sale_price'])
```

```python
num_col = [i for i in data.select_dtypes(include=["float64"]).columns if i != "rating"]
```

## IQR

```python
for i in num_col:
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    IQR = Q3-Q1
    UL = Q3 + 1.5 *IQR
    LL = Q1 - 1.5*IQR
    data[i] = np.where(data[i]>UL,UL,
                       np.where(data[i]<LL,LL,
                                data[i]))
```

```python
plt.figure(figsize=(9,5))
sns.boxplot(data)
```
