1. Handling Missing Values

```python
print("Any missing Values in our data set?", df.isnull().values.any())
```

```python
df.isnull().sum()
```

```python
#to visualize missing values
sns.heatmap(df.isnull())
```

Missing values are shown in "Revenue(Millions)" and "Metascore" columns with Lighter shade.

```python
#percentage of missing values
per_missing=(df.isnull().sum()*100)/len(df)
per_missing
```

```python
sns.boxplot(df['Revenue (Millions)'])
```

```python
#since revenue column contains outliers as well, we are replacing missing values with "median"
df["Revenue (Millions)"].fillna(df["Revenue (Millions)"].median(), inplace=True)
```

```python
sns.boxplot(df["Metascore"])
```

```python
#Metascore column has no outlier value so replacing with Mean
df["Metascore"].fillna(df["Metascore"].mean(), inplace=True)
```

20. Classify Movies Based on Ratings[Excellent, Good and Average]

```python
def rating(rating):
  if rating>-7.0:
    return "Excellent"
  elif rating>-6.0:
    return "Good"
  else:
    return "Average"
```

22.Find Unique values from Genre Column

```python
df['Genre']
```

```python
#as there are comma separated values
#we will first separate values and feed them into a list
list1=[]
for value in df['Genre']:
  list1.append(value.split(','))
```

```python
list1
```

```python
#convert 2d list into 1 list
one_d=[]
for item in list1:
  for item1 in item:
    one_d.append(item1)
```

```python
one_d
```

```python
#find unique value in this genre
uni_list=[]
for item in one_d:
  if item not in uni_list:
    uni_list.append(item)
```

```python
len(uni_list)
```
