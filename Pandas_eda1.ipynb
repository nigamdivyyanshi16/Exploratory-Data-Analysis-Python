Q10. FIND TOTAL NUMBER OF STUDENTS HAVING MARKS BETWEEN 90 TO 100(INCLUSIVE) USING BETWEEN METHOD

```python
#w/o between
df[df['Marks']>=90]
#generatees boolean series
#when passed inside df then it matches true value and returns rows
```

```python
len(df[(df['Marks']>=90) &  (df['Marks']<=100)])
#when more than 2 or 2 condiition always inside paranthesis
```

```python
#total 4 students have marks between 90 to 100
```

```python
#with between
sum(df['Marks'].between(90,100))
#sum will calculate total number of time True came from expression inside
```

Q12. APPLY METHOD

```python
def marks(x):
  return x//2
```

```python
#apply function
df['Marks'].apply(marks)
#pass the function without paranthesis
#this will make changes to existing
```

```python
#to make changes by adding a new column
df['Half Marks']=df['Marks'].apply(marks)
df
```

```python
#apply lambda function
df['Marks'].apply(lambda x:x//2)
```
